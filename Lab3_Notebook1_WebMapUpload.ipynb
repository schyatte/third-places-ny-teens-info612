In this third lab, we're going to upload the maps made previously to ArcGIS Online. Web maps are great for sharing projects with the wider public, and publishing on ArcGIS online is an easy way to upload an interactive map for people to explore. It's generally a pretty simple process, but sometimes there are formatting issues between the desktop and online map environments.

For example, selection layers are not supported in web maps, so we need to copy the features of the Third places Within Walking Distance layer to a new layer.

```python
arcpy.management.CopyFeatures(
    in_features="Third Places Within Walking Distance",
    out_feature_class=r"G:\My Drive\INFO612_Advanced_GIS\Lab2_NYTeens_Analysis\Lab2_NYTeens_Analysis.gdb\ThirdPlaces_Features",
    config_keyword="",
    spatial_grid_1=None,
    spatial_grid_2=None,
    spatial_grid_3=None
)
```

Remove the original Third Places Within Walking Distance from the map. Uncheck it from the contents pane, but don't remove it from the map, so that it can be viewed in the web map.
